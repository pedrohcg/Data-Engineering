```
%sql
create table users (
  user_id int,
  user_fname string,
  user_lname string,
  user_phones array<struct<phone_type: string, phone_number: string>>
)
```

```
%sql
insert into users values (1, 'Scott', 'Tiger', array(struct('home', '+1 (234) 567 8901'), struct('mobile', '+1 (123) 456 7890'))),
(2, 'Donald', 'Duck', null),
(3, 'Mickey', 'Mouse', array(struct('home', '+1 (123) 456 9012')))
```

```
%sql
select * from users
```

```
%sql
select user_id, explode(user_phones) user_phone
from users
```

```
%sql
select user_id, explode_outer(user_phones) user_phone
from users
```

```
%sql
with users_phones_exploded as (
  select user_id, explode_outer(user_phones) user_phone
  from users
) 
select user_id, user_phone.* from users_phones_exploded
```

```
%sql
create table users (
  user_id int,
  user_fname string,
  user_lanme string,
  user_phone_type string,
  user_phone_number string
)
```

```
%sql
insert into users values (1, 'Scott', 'Tiger', 'home', '+1 (234) 567 8901'),
(1, 'Scott', 'Tiger', 'mobile', '+1 (234) 456 7890'), 
(2, 'Donald', 'Duck', null, null),
(3, 'Mickey', 'Mouse', 'home', '+1 (123) 456 9012')
```

```
%sql
select * from users
```

```
%sql
select user_id, user_fname, user_lanme, collect_list(user_phone_number) user_phones
from users
group by 1, 2, 3
```

```
%sql
select user_id, user_fname, user_lanme, concat_ws(',', collect_list(user_phone_number)) user_phones
from users
group by 1, 2, 3
```

```
%sql
select user_id, user_fname, user_lanme, struct(user_phone_type, user_phone_number) user_phone
from users
```

```
%sql
select user_id, user_fname, user_lanme, nvl2(user_phone_number, struct(user_phone_type, user_phone_number), null) user_phone
from users
```

```
%sql
select user_id, user_fname, user_lanme, collect_list(nvl2(user_phone_number, struct(user_phone_type, user_phone_number), null)) user_phone
from users
group by 1, 2, 3
```
