Get revenue for each order_id
// 
if order_items does not have item details associated with a given order id, then revenue should be 0

```
%sql
select order_id, order_date, o.order_customer_id, o.order_status, coalesce(round(sum(order_item_subtotal), 2), 0) order_revenue
from order_items oi
right join orders o
on oi.order_item_order_id = o.order_id
group by 1, 2, 3, 4 
order by 1
```

Get revenue for each order_id where orders are placed in 2014 january and the status is either COMPLETE or CLOSED
// 
if order_items does not have item details associated with a given order id, then revenue should be 0
//
along with details such as order customer id, status, date, etc

```
%sql
select order_id, order_date, o.order_customer_id, o.order_status, coalesce(round(sum(order_item_subtotal), 2), 0) order_revenue
from order_items oi
right join orders o
on oi.order_item_order_id = o.order_id
where o.order_status in ('COMPLETE', 'CLOSED')
and year(o.order_date) = 2014
and month(o.order_date) = 1 
group by 1, 2, 3, 4 
order by 1
limit 10
```

Get revenue for each order_id where orders are placed in 2014 january and the status is either COMPLETE or CLOSED
// 
if order_items does not have item details associated with a given order id, then revenue should be 0
//
along with details such as order customer id, status, date, etc
//
get only those orders where the revenue is grater than 500

```
%sql
select order_id, order_date, o.order_customer_id, o.order_status, coalesce(round(sum(order_item_subtotal), 2), 0) order_revenue
from order_items oi
right join orders o
on oi.order_item_order_id = o.order_id
where o.order_status in ('COMPLETE', 'CLOSED')
and year(o.order_date) = 2014
and month(o.order_date) = 1 
group by 1, 2, 3, 4
having order_revenue > 500 
order by 1
limit 10
```

```
%sql
merge into daily_revenue t 
using (
  select o.order_date, round(sum(oi.order_item_subtotal), 2) order_revenue
  from orders o
  join order_items oi
  on o.order_id = oi.order_item_order_id
  where o.order_status in ('COMPLETE', 'CLOSED')
  and year(o.order_date) = 2013
  and month(o.order_date) = 8
  group by 1
) s
on s.order_date = t.order_date
when matched then update set *
when not matched then insert *
```
